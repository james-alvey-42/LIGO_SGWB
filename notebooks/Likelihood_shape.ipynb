```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xi = np.linspace(0, 1, 100)
```

```python
N_segs = [3, 10, 30, 100, 300]

fig, ax = plt.subplots(1, len(N_segs), figsize=(12, 3))

for n in range(len(N_segs)):
    a = np.random.uniform(0., .8, N_segs[n])
    b = 1 - a #np.random.uniform(0, 1, N_segs[n])
    lik = np.exp( np.sum( np.log(xi[:,None] * a[None,:] + (1 - xi[:,None]) * b[None,:] ), axis=-1) )
    ax[n].plot(xi, lik)

plt.tight_layout()
```
